A continuous random variable with cumulative distribution function $F$ has the median value $m$ such that $F(m) = 0.5$. That is, a random variable is just as likely to be larger than its median as it is to be smaller. A continuous random variable with density $f$ has the mode value $x$ for which $f(x)$ attains its maximum. For each of the following three random variables, $(i)$ state the density function, $(ii)$ compute the median, mode and mean for the random variable, and $(iii)$ Provide at least one graph for the density function using values of the parameter(s) that you select.  Indicate the median, mode, and mean values on your graph.  (The purpose of this problem is to see the relative locations of the median, mode, and mean for the different random variables).

**a)** $W$ which is uniformly distributed over the interval $[a, b]$, for some value $a,b \in \mathbb{R}$.

```r
# Plot a uniform density from a to b
# In this case, we will use 0 to 1 as the the interval
x = seq(0, 1, 0.05)
y = dunif(x)
# plot the uniform distribution
plot(x=x, y=y, type="l", col='black')
# Median is at 1/(a+b), which is 1/(1+0)=1/2 for our plot
median = median(x)
# Draw vertical median line
abline(v=median, col="red")
# Mean is at 1/2 as well
mean = median(x)
# Draw vertical mean line
# The blue line will overlap the red line
abline(v=mean, col="blue")
# The mode of a continuous random variable corresponds to the x value(s) 
# at which the probability density function reaches a local maximum, or a peak. 
# In uniform distribution, there is no mode, as all values of f(x) are equal.
```

$W$ is uniform on [a,b] Its density function is $f_W(w) = 1/(b-a)$ for all $a \le w \le b$ and $f_W(w) = 0$ for all other values of $w$. 
* Mean $E[W] = (a+b)/2$. 
* Median for uniform is (a+b)/2 as well. 
* In uniform distribution, there is no mode, as all values of f(x) are equal.

**b)** $X$ which is normal with parameters $\mu$ and $\sigma^2$, for some value $\mu, \sigma^2 \in \mathbb{R}$.

```r
# Simulate a standard normal distribution
x = seq(-5, 5, 0.1)
y = dnorm(x)
# Plot the standard normal
plot(x=x, y=y, type="l")
# The mean, median, and mode of a standard normal distribution are equal to zero. 
median = 0
mean = 0
mode = 0
abline(v=median, col="red")
abline(v=mean, col="blue")
abline(v=mode, col="green")
```

For $X \sim N(\mu, \sigma^2)$, the median=mean=mode=$\mu$. When it is a standard normal deviation, they are all equals to 0.

**c)** $Y$ which is exponential with rate $\lambda \in \mathbb{R}$.

```r
# Your Code Here
# Generate an exponential distribution with rate lambda
# For our case, let lambda = 1
lambda = 1
x = seq(0, 20, 0.1)
y = dexp(x, rate=lambda)
# Plot the exponential distribution
plot(x=x, y=y, type="l")

# Mean is 1/lambda.
mean = 1 / lambda
abline(v=mean, col="blue")

# Median is ln(2)/lambda.
median = log(2)/lambda
abline(v=median, col="red")

# Mode is at x=0.
mode = 0
abline(v=mode, col="green")
```

$Y$ is exponential with rate $\lambda$ so it has density function $f_Y(y)=\lambda e^{-\lambda y}$ for $y \ge 0$ and $f_Y(y)=0$ for $y<0$.
* The mean for exponential distribution is $E[Y] = 1/\lambda$.  (blue colored)
* Median of the ex[onential distribution is $m = (\ln 2)/\lambda$. (red colored)
* The mode of a continuous random variable is not the point where its probability is most massive.   It is where it is most dense. This is at its maximum when x=0.   Hence the mode of the exponential distribution is: 0.(green colored)

**Part c)**

Suppose we are only able to sample from the standard normal distribution $N(0,1)$. Could we take those samples and perform a simple transformation so that they're samples from $N(2,3)$? Try this, and plot another histogram of the transformed data, again with the same axes. Does your histogram based of the transformed data look like the histogram from **Part B**?

```r
# Your Code Here 
# Since 
# (standard.normal.dist.samples.3 - mean) / sigma = standard.normal.dist.samples.1
# standard.normal.dist.samples.3 = standard.normal.dist.samples.1 * sigma + mu

standard.normal.dist.samples.3 = standard.normal.dist.samples.1 * sqrt(3) + 2
hist(standard.normal.dist.samples.3, freq=FALSE, xlim=c(-10, 10), ylim=c(0, 0.5))
abline(v=2,col='red')
```

Two histograms do match, as they have the same mean and variance ( or standard deviation)
