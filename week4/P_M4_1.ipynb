**a)** $W$ which is uniformly distributed over the interval $[a, b]$, for some value $a,b \in \mathbb{R}$.

```r
# Plot a uniform density from a to b
# In this case, we will use 0 to 1 as the the interval
x = seq(0, 1, 0.05)
y = dunif(x)
# plot the uniform distribution
plot(x=x, y=y, type="l", col='purple')
# Median is at 1/(a+b), which is 1/(1+0)=1/2 for our plot
median = median(x)
# Draw vertical median line
abline(v=median, col="red")
# Mean is at 1/2 as well
mean = median(x)
# Draw vertical mean line
# The blue line will overlap the red line
abline(v=mean, col="blue")
# The mode of a continuous random variable corresponds to the x value(s) 
# at which the probability density function reaches a local maximum, or a peak. 
# In uniform distribution, there is no mode, as all values of f(x) are equal.
```

$W$ is uniform on [a,b] Its density function is $f_W(w) = 1/(b-a)$ for all $a \le w \le b$ and $f_W(w) = 0$ for all other values of $w$. 
* Mean $E[W] = (a+b)/2$. 
* Median for uniform is (a+b)/2 as well. 
* In uniform distribution, there is no mode, as all values of f(x) are equal.

**b)** $X$ which is normal with parameters $\mu$ and $\sigma^2$, for some value $\mu, \sigma^2 \in \mathbb{R}$.

```r
# Simulate a standard normal distribution
x = seq(-5, 5, 0.1)
y = dnorm(x)
# Plot the standard normal
plot(x=x, y=y, type="l")
# The mean, median, and mode of a standard normal distribution are equal to zero. 
median = 0
mean = 0
mode = 0
abline(v=median, col="red")
abline(v=mean, col="blue")
abline(v=mode, col="green")
```

For $X \sim N(\mu, \sigma^2)$, the median=mean=mode=$\mu$. When it is a standard normal deviation, they are all equals to 0.
